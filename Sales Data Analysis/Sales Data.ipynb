```python
sv1=sv.rename(columns={'Unnamed: 0': 'Serial No'})
sv1.head()
```

```python
sv1.isnull().sum()
```

```python
sv1['Price Each']=sv1['Price Each'].fillna(0)
sv1['Month']=sv1['Month'].fillna(0)
sv1['Sales']=sv1['Sales'].fillna(0)
sv1.head()
```

```python
sv1.duplicated()
```

```python
sv1.drop_duplicates()
```

```python
top=sv1.nlargest(10,'Sales')
print("Top Sales:\n",top[['Product','Sales']])
```

```python
low=sv1.nsmallest(10,'Sales')
print("Low Sales:\n",low[['Product','Sales']])
```

```python
ssv.figure(figsize=(10,6))
sizes=top["Quantity Ordered"]
labels=top["City"]
ssv.pie(sizes,labels=labels,autopct="%2.1f")
ssv.show()
```

```python
ssv.figure(figsize=(13,10))
ssv.plot(sv1['Sales'],sv1['Price Each'],color='pink',linestyle='-')
ssv.grid(True)
ssv.show()
```

```python
ssv.figure(figsize=(13,5))
ssv.bar(sv1['Quantity Ordered'],sv1['Sales'],color='skyblue')
ssv.show()
```
