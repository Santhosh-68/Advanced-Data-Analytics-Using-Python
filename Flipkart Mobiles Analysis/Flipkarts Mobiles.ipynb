```python
sv['Memory']=sv['Memory'].fillna(0)
sv['Storage']=sv['Storage'].fillna(0)
sv['Rating']=sv['Rating'].fillna(0)
sv.head()
```

```python
sv.duplicated()
```

```python
sv=sv.drop_duplicates()
```

```python
sv.head()
```

```python
sv.tail()
```

```python
top=sv.nlargest(10,'Rating')
print("Top Prices:\n",top[['Brand','Selling Price']])
```

```python
bottom=sv.nsmallest(10,'Rating')
print("Lowest Price:\n",bottom[['Brand','Selling Price']])
```

```python
high=sv.nlargest(10,'Rating')
print("Top Brands:\n",high[['Brand','Model','Color']])
```

```python
low=sv.nsmallest(10,'Rating')
print("Lowest Brands:\n",low[['Brand','Model','Color']])
```

```python
ssv.figure(figsize=(5,5))
sizes=high["Original Price"]
labels=high["Brand"]
ssv.pie(sizes,labels=labels,autopct="%2.1f")
ssv.show()
```

```python
ssv.figure(figsize=(13,5))
ssv.bar(sv['Brand'],sv['Selling Price'],color='indigo')
```

```python
ssv.figure(figsize=(13,10))
ssv.plot(sv['Brand'],sv['Original Price'],color='green',linestyle='-')
ssv.grid(True)
ssv.show()
```
