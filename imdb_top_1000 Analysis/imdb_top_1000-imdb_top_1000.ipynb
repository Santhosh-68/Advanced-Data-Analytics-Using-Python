```python
sv['Gross']=sv['Gross'].str.replace(',',"").astype(float)
```

```python
sv['Meta_score'].fillna(sv['Meta_score'].mean(),inplace=True)
sv['Certificate'].fillna('Not specified',inplace=True)
sv['Gross'].fillna(sv['Gross'].mean(),inplace=True)
```

```python
duplicate_rows=sv.duplicated()
print("Duplicate Rows:\n",duplicate_rows)

#Remove Duplicate Rows
sv=sv.drop_duplicates()
```

```python
sv[sv.duplicated()]
```

```python
correlation = sv['IMDB_Rating'].corr(sv['Gross'])
print(correlation)
```

```python
# Top Scores
top_scores=sv.nlargest(10,'Meta_score')
print("Top Scores:\n",top_scores[['Meta_score','Runtime']])
```

```python
# Low Scores
low_scores=sv.nsmallest(10,'Meta_score')
print("Low Scores:\n",low_scores[['Meta_score','Runtime']])
```

```python
# Genre difference
genre_difference=sv['Genre'].value_counts()
print("Genre difference")

print(genre_difference)
```

```python
avg_gross_genre = sv.groupby('Genre')['Gross'].mean()
top_genre_gross = avg_gross_genre.idxmax()
top_genre_avg_gross = avg_gross_genre.max()
print(top_genre_gross)
print(top_genre_avg_gross)
```

```python
director_avg_rating = sv.groupby('Director')['IMDB_Rating'].mean()
best_director = director_avg_rating.idxmax()
print(best_director)
```

```python
avg_rating_yearly = sv.groupby('Released_Year')['IMDB_Rating'].mean()
print(avg_rating_yearly)
```

```python
max_rating_year = avg_rating_yearly.idxmax()
max_rating_value = avg_rating_yearly.max()
print(max_rating_year)
print(max_rating_value)
```

```python
total_revenue_yearly = sv.groupby('Released_Year')['Gross'].sum()
print(total_revenue_yearly)
```

```python
max_rating_year = total_revenue_yearly.idxmax()
max_rating_value = total_revenue_yearly.max()
print(max_rating_year)
print(max_rating_value)
```

```python
top_directors_plot = director_avg_rating.nlargest(7).plot(kind='bar', xlabel='Director', ylabel='Average IMDb Rating',
                                                          title='Top-rated Directors',color='pink')
ssv.show()
```

```python
top_genres_ratings = genre_difference.nlargest(5)
ssv.figure(figsize=(10, 8))
ssv.pie(top_genres_ratings, labels=top_genres_ratings.index, autopct='%1.1f%%', startangle=90, colors=ssv.cm.Paired.colors)
ssv.title('Top Genres Based on IMDb Ratings')
ssv.show()
```
