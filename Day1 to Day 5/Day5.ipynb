```python
ssv1=ssv.drop(['race','gender'],axis=1)
ssv1.head()
```

```python
ssv1['Total']=ssv1['Maths']+ssv1['English']+ssv1['History']+ssv1['computer']+ssv1['Science']
ssv1.head()
```

```python
ssv2=ssv1.drop(['percentage'],axis=1)
ssv2.head()
```

```python
ssv2['percentage']=ssv2['Total']*100/500
ssv2.head()
```
